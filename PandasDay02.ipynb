```python
import pandas as pd
```

```python
# create new data
employees = pd.DataFrame({
    'emp_id' : [1,2,3,4],
    'name' : ['Ramesh','Rajesh','Rabin','Ram'],
    'dept_id' : [10,20,30,40]
})
```

```python
# create new data
departments = pd.DataFrame({
    'dept_id' : [10,20,40],
    'dept_name' : ['HR','IT','Finance']
})
```

```python
employees
```

```python
departments
```

```python
# Inner - only matching rows from both tables
pd.merge(employees, departments, on = 'dept_id', how = 'inner')
```

```python
# Left - all rows from left, matched rows from right (NaN if no match)
pd.merge(employees, departments, on = 'dept_id', how = 'left')
```

```python
# Right - all rows from right table
pd.merge(employees, departments, on = 'dept_id', how = 'right')
```

```python
# Outer - all rows from both tables
pd.merge(employees, departments, on = 'dept_id', how = 'outer')
```
