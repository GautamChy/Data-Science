# Types of Statistics

 1 ) Descriptive Statistics
 - Definion:  It means giving information only about the data that is available.( It means Jey data chhha tyesko barey ma matra information dinney)
 - Academic version: It refers to summarizing and describing only the data we have.
   * It consists of organizing and summerizing data.
   * Measure of Central Tendency(Mean, Median and Mode)
   * Measure of Dispersion(Variance and Standard Deviation)
   * Diffferent types of distribution of data(Histogram,pdf,pmf)

   Note: People those who work as Data analysis they used this Descriptive Statistics

2 ) Inferential Statistics

- Definion: It means taking the available data and using it to make predictions or draw conclusions. (Haami sitaw vaako data liney and tyelai predict or form conclusion garney.)
- Academic version: It involves using sample data to predict, estimate, or draw conclusions about a larger population.
  * It consists of using data that we have collected to form conclusions( conclusion means predict). (Use sample data to draw conclusion(predict) for population)
  * FOR EXAMPLE: euta election vaairako thau ko sample linney and tyevako population predict garney.(Use sample data to draw conclusion(predict) for population)
  * Z test, t test, Chi square test(Hypothesis Testing)

Note: People those who work as Data science they used this Inferential Statistics.

# Measure of Dispersion
* Measure of dispersion => Measures of dispersion quantify how spread out or scattered a set of data points is around a central value (like the mean or median). They show the variability or consistency within the data.

* **variance**
  - jati dherai varience ko value hunca tyeti dherai data spread vako hunca.

  - Sample => It means less data.

  - Sample variance
    - x = Data point,x^- = Sample Mean, n = sample size
    - Why we divide sample varience by n -1 ?
    - It is done so that we can create unbiased estimator of the population variance.

# Skewness

Skewness measures whether a distribution is symmetric or lopsided.

1. Symmetric Distribution (Skewness ≈ 0)
* Left and right sides are roughly mirror images.
* Mean ≈ Median ≈ Mode

Example: Heights of adults often look approximately symmetric.



**2. Positive (Right) Skew**
Screenshot 2026-06-03 at 11.51.51 AM.png

* Long tail extends to the right.
* A few very large values pull the mean upward.
* Mean > Median > Mode

Example:

* Household income
* House prices

Most people earn moderate incomes, while a small number earn very high incomes.

* Left side tiraw thulo chucho vayeraw right side tiraw tail extend vako hunxa positive(Right)Skew ma.

 **3. Negative (Left) Skew**
 Screenshot 2026-06-03 at 11.54.07 AM.png

* Long tail extends to the left.
* A few very small values pull the mean downward.
* Mean < Median < Mode

Example:

Easy exam scores where most students score high.

* Right side tiraw thulo chucho vayeraw left side tiraw tail extend vako hunxa Negative(Left)Skew ma.
