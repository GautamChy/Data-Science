# Min-max scaling(rescales to a range, often 0–1) = It is  mainly used in Deep learning

*FORMULA
* x' = x - xmin / xmax - xmin

# Feature scaling
- Min-max scaling, z-score standardization, and normalization are generally considered feature scaling (a preprocessing step).

Normalization[Min Max Scaler] --> 0 to 1

Avg 24,25,26,27,28,30

Transformation --> xscaled = (x - xmin)/(xmax - xmin)

(24-24)/(30-24)

Mostly use standardization in Machine learning and normalization in deep learning.

```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
df.head()
```

```python
# importing MinMaxScaler from sklearn.preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max
```

```python
df[['distance', 'fare', 'tip']]
```

```python
# Checking the max and min values of the columns. This is the original data
max(df['distance']), min(df['distance']), max(df['fare']), max(df['tip']),min(df['tip'])
```

```python
# fitting the data to the MinMaxScaler
min_max.fit(df[['distance', 'fare', 'tip']])
```

```python
# transform the data to the MinMaxScaler
min_max.transform(df[['distance', 'fare', 'tip']])
```

```python
# fitting and transform the data to the MinMaxScaler, we did both fitting and transform at once.
min_max.fit_transform(df[['distance', 'fare', 'tip']])
```

```python
 # take above array data which is in 2D array, and take original columns name
import pandas as pd
df = pd.DataFrame(min_max.fit_transform(df[['distance', 'fare', 'tip']]), columns=['distance', 'fare', 'tip'])
df
```

```python
df.head()
```

```python
# Checking the max and min values of the columns. This is after Normalization data
max(df['distance']), min(df['distance']), max(df['fare']), max(df['tip']),min(df['tip'])
```
