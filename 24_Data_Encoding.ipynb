# Advantages
* One hot encoding convert categorical data(character) into numerical data which is always in 1 and 0

# Disadvantages
* It Create spare matrix.(10 locations only takes 1 and 0 which lead to overfitting) ( Search on google spare matrix)
* 1000 categories (100 features). No of features increased, performance of model decreases
* i.e features means columns

```python
# import pandas and one hot encoder
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# create a dataframe with a single column of color categories
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df
```

```python
type(df)
```

```python
df.dtypes
```

According to the above color column all data is in object that is categorical form

```python
## Create an instance of the OneHotEncoder,Note: OneHotEncoder() is a class. Hold on OneHotEncoder then see
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(df[['color']])
```

Compressed Sparse Row sparse matrix of dtype 'float64'
	with 6 stored elements and shape (6, 3)>
  * It means spare matrix banyo where 6 ota rows and 3 ota columns banyo.

```python
# convert spare matrix into array by using toaraay()
ohe.fit_transform(df[['color']]).toarray()
```

```python
# fit the encoder to the dataframe and transform the category
encoded = ohe.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python
# take above encoded data which is in 2D array, and take columns name
encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['color'])) # get_feature_names_out is a function.Hold on get_feature_names_out                        )
```

```python
encoded_df  # column name is always in aplhabetical order i.e a,b,c so blue,green,red. Here,we can see now categorical(character is in numerical)
```

```python
df
```

```python
# By default NaN while concat
pd.concat([df, encoded_df])
```

```python
# Removing NaN above data by puting axis = 1
pd.concat([df, encoded_df], axis=1)
```
