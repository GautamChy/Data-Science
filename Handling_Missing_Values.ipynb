# fillna()
* it’s used to replace missing values (NaN) instead of removing them.
* It fills in empty/missing data with a value you choose.

```python
# filling average age value in new column i.e 'Age_mean' is new column
df['Age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
# Original age column → stays the same (still has NaNs)
# New Age_mean column → has no missing values
df[['Age_mean', 'age']]   #Age_mean column there are added mean of age value but stil NaN in age column
```

```python
# Checking missing values in Age_mean column
df['Age_mean'].isnull().sum() # There are zero missing value i.e NaN
```

```python
# Checking missing values in Age column
df['age'].isnull().sum()  # There are 177 missing values
```

# Median Value Implutation

```python
# Median Value Implutation
```

```python
df['age'].median()
```

```python
# filling median age value in new column i.e 'Age_median'
df['Age_median'] = df['age'].fillna(df['age'].median())
```

```python
df
```

```python
df.dropna().shape
```

```python
# Original age column → stays the same (still has NaNs)
# New Age_median column → has no missing values
df[['Age_median', 'Age_mean', 'age']]
```

```python
# Checking missing values
df['Age_median'].isnull().sum() # There are no missing values
```
