```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'salary' : np.random.randint(30000, 120000, 100000)
})
```

```python
df.shape
```

```python
df.head()
```

```python
# This way takes more time
result = []
for val in df['salary']:
  result.append(val * 1.1)
df['new_salary'] = result      # new_salary means new column store a value by result
```

```python
# Vectorized opearation
# This way take less time
# Arithmetic
df['tax'] = df['salary'] * 0.25    # tax is new column, salary is old column so salary * 0.25
df['bonus'] = df['salary'] * 0.1
df['net']   = df['salary'] - df['tax']
df['ratio'] =df['salary'] / df['bonus']
```
