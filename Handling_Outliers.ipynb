# Finding Outliers

Screenshot 2026-05-21 at 9.33.45 PM.png

[Lower Fence <---> Higher Fence]

```python
# Calculate the five number summary, 0 means minimum value, 0.25 means 25 percentile, 0.50 means median or 50 percentile, 0.75 means 75 percentile, 1 means maximum percentile.
minimum, Q1, median, Q3, maximum = np.quantile(numbers, [0, 0.25, 0.5, 0.75, 1])
```

```python
maximum, Q3, Q1
```

```python
minimum, median
```

```python
# finding 50 percentile by using formula
iqr = Q3 - Q1
print(iqr)
```

```python
lower_fence = Q1 - 1.5 * iqr
higher_fence = Q3 + 1.5 * iqr
```

```python
lower_fence, higher_fence # If any data between 0.75 to 142.75 is called normal data.
```

```python
numbers
```

```python
outliers = []
for num in numbers:
  if num < lower_fence or num > higher_fence:
    outliers.append(num)
```

```python
print(outliers)
```
